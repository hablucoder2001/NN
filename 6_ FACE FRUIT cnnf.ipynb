```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. Folder paths
train_path = r"D:\4TH YEAR 2ND SEMESTER\ICE-4206 (Neural Networks Sessionals)\FOR LAB EXAM PREP\BIRD DATASET\bird"
val_path   = r"D:\4TH YEAR 2ND SEMESTER\ICE-4206 (Neural Networks Sessionals)\FOR LAB EXAM PREP\BIRD DATASET\val-bird"
test_path  = r"D:\4TH YEAR 2ND SEMESTER\ICE-4206 (Neural Networks Sessionals)\FOR LAB EXAM PREP\BIRD DATASET\test_bird"

# 2. Load datasets
train_data = tf.keras.utils.image_dataset_from_directory(
    train_path,
    image_size=(128, 128),
    batch_size=32
)

val_data = tf.keras.utils.image_dataset_from_directory(
    val_path,
    image_size=(128, 128),
    batch_size=32
)

test_data = tf.keras.utils.image_dataset_from_directory(
    test_path,
    image_size=(128, 128),
    batch_size=32
)

print("Classes:", train_data.class_names)

# 3. Simple CNN model
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(128, 128, 3)),

    layers.Conv2D(16, 3, activation="relu"),
    layers.MaxPooling2D(),

    layers.Conv2D(32, 3, activation="relu"),
    layers.MaxPooling2D(),

    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(3, activation="softmax")
])

# 4. Compile model
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# 5. Train model
model.fit(
    train_data,
    validation_data=val_data,
    epochs=5
)

# 6. Test model
loss, accuracy = model.evaluate(test_data)

print("Test Accuracy:", accuracy)
```
